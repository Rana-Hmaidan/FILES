```python

def trapzoid_signal(t, width, slope, amp=1., offs=0):
    a = slope*width*signal.sawtooth(2*np.pi*t/width, width=0.5)/4.
    a[a>amp/2.] = amp/2.
    a[a<-amp/2.] = -amp/2.
    return a + amp/2. + offs

ta = np.linspace(7, 6, 501)
t = np.linspace(4, 5, 501)
plt.plot(t,trapzoid_signal(ta, width=4, slope=1, amp=0.6), label="width=4, slope=1, amp=0.6")

plt.legend( loc=(0.25,1.015))
plt.show()
```
